```python
import os
import numpy as np
import pandas as pd
```

```python
def split_into_sequences(df,save_path):
    i = 0
    while i<(len(df)):
        dfs = df.iloc[i:i+10.:]
        dfs.to_csv(save_path+str(int(i/10))+".csv")
        i = i+10
    dfs = df.iloc[i:len(df),:]
    dfs.to_csv(save_path+str(int(i/10))+".csv")
```
